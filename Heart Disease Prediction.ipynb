# Data Preprocessintg

# Convert categorical variables into dummy_var

```python
data.info()
```

```python
dataset = pd.get_dummies(data,columns=['sex', 'cp','fbs','restecg','exang', 'slope', 'ca', 'thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
Std = StandardScaler()
columns_to_scale = ['age','trestbps','chol','thalach','oldpeak']
dataset[columns_to_scale] = Std.fit_transform(dataset[columns_to_scale])
dataset.head()
```

# Declare Feture var and target var

```python
y = dataset['target']
X = dataset.drop(['target'],axis=1)
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
  knn_classifier = KNeighborsClassifier(n_neighbors=k)
  score = cross_val_score(knn_classifier,X,y,cv=10)
  knn_scores.append(score.mean()) 
```

```python
fig =plt.figure(figsize=(20,15))
plt.plot([k for k in range(1,21)],knn_scores,color='red')
for i in range(1,21):
  plt.text(i,knn_scores[i-1],(i,knn_scores[i-1]))
plt.xticks([i for i in range(1,21)])
plt.xlabel('Nuber of Neighbours (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifiers Scores for Different K values')
plt.show()
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=12)
score = cross_val_score(knn_classifier,X,y,cv=10)
score.mean()
```

# Random forest classifiers

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=10)
score = cross_val_score(RF,X,y,cv=10) 
score.mean()
```
