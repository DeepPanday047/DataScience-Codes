```python
import tensorflow as tf
from keras.applications.vgg16 import VGG16
import os,cv2
from keras.models import Model
from keras.layers import *
from keras.preprocessing.image import ImageDataGenerator
```

```python
vgg16 = VGG16(include_top=False,weights='imagenet',input_shape=(224,224,3))
```

```python
vgg16.summary()
```

```python
for layer in vgg16.layers:
  layer.trainable = False # Freeze all vgg16 layers
  print(layer.name, layer.trainable)
```

```python
flat = tf.keras.layers.Flatten() (vgg16.output)
dense = tf.keras.layers.Dense(64, activation='relu') (flat)
output = tf.keras.layers.Dense(120, activation="softmax") (dense)
```

```python
final_model = tf.keras.models.Model(inputs=[vgg16.input], outputs=[output])
```

```python
final_model.summary()
```

```python
for layer in final_model.layers:
  print(layer.name, layer.trainable)
```
