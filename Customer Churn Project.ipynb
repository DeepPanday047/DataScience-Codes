# Modeling

 # Linear regression

```python
data.columns
```

```python
X = data.iloc[:,5:6]
y = data['MonthlyCharges']
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,y_train)
```

```python
y_pred_LR= LR.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_test, y_pred_LR)
print(f'Root Mean Squared Error: {rmse}')
```

```python
error = y_test - y_pred_LR
```

# Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
```

```python
X = data[['MonthlyCharges']]
y = data['Churn']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.35,random_state=42)
```

```python
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
y_pred_LOR = LR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
acc = accuracy_score(y_test,y_pred_LOR)
print('\n Accuracy Score:',acc)
con = confusion_matrix(y_test,y_pred_LOR)
print('\n Confusion Matrix\n',con)

```

# Random Forest
