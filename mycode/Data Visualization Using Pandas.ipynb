Histogram

```python
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
data = pd.DataFrame(data=iris.data,columns=['sepal length (cm)','sepal width (cm)','petal length (cm)', 'petal width (cm)'])
```

```python
data
```

```python
data.plot.hist(alpha =.4)
plt.show()
```

BoxPlot 

```python
data.plot.box()
plt.show()
```

To Plot Horizontally

```python
data.plot.box(vert=False)
```

Area Charts

```python
dff = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
```

```python
dff.plot.area()
plt.show()
```
